```python
import numpy as np
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt
from statistics import mean
import seaborn as sns
```

```python
# mangaid= 1-217041
#f= [VLSS,TGSS,WENSS, NVSS,FIRST]

x= [74,150,325,1400,3500 ]
y=[5684,2766,1621,272,114]
e= [10,276,10,1, 5] 

plt.errorbar(x,y, yerr=e, linestyle='dashed',mfc='b',marker='o',ecolor='r', barsabove=True, capsize=10,elinewidth=3 )

sns.regplot(x,y,ci=None)
plt.xlabel(r'${\rm Frequency }\,[$MHz$]$', fontsize='x-large')
plt.ylabel(r'${\rm Flux Density}\,[Log(milli Jy)]$', fontsize='x-large')
```

```python
# mangaid= 1-217041

from scipy.stats import linregress

#f= [VLSS,TGSS,WENSS, NVSS,FIRST]

x= [74,150,325,1400,1400 ]
y=[5684,2766,1621,272,114]
e= [90,276,10,9.6, 0.5] 

linregress= linregress(x, y)

print(linregress)

logA = np.log(x) 
logB = np.log(y)


plt.errorbar(x,y, yerr=e, mfc='k',linestyle='dashed',marker='o',markersize=8,ecolor='deepskyblue',barsabove=True, capsize=20,elinewidth=4 )

#plt.loglog(x, y,linestyle='dashed')


#create scatterplot with regression line
m, c = np.polyfit(logA, logB, 1) # fit log(y) = m*log(x) + c
y_fit = np.exp(m*logA + c) 

plt.plot(x, y,color='k')
plt.plot(x, y_fit, color='r',ls='--')
plt.yscale('symlog')
plt.xscale('symlog')
plt.xlabel(' Frequency log[MHz]', fontsize='x-large')
plt.ylabel(' Flux Density log[milli Jy]', fontsize='x-large')

#sns.regplot(x,y,ci=None)
#data = pd.DataFrame(data={'x': x, 'y': y})
##f= plt.subplots(figsize=(7, 7))
#ax.set(xscale="log", yscale="log")
```

```python
# mangaid= 1-217041
#f= [VLSS,TGSS,WENSS, NVSS,FIRST]
x= [74,150,325,1400,1400 ]
y=[5684,2766,1621,272,114]
e= [90,276,10,9.6, 0.5] 
logA = np.log(x) 
logB = np.log(y)

plt.errorbar(x,y, yerr=e, mfc='aqua',linestyle='dashed',marker='o',markersize=8,ecolor='m',barsabove=True, capsize=20,elinewidth=4 )

#plt.loglog(x, y,linestyle='dashed')
#create scatterplot with regression line
m, c = np.polyfit(logA, logB, 1) # fit log(y) = m*log(x) + c
y_fit = np.exp(m*logA + c) 
plt.plot(x, y,color='k',ls='--')
plt.plot(x, y_fit, color='r')

plt.yscale('symlog')
plt.xscale('symlog')

#sns.regplot(x,y,ci=None)
#data = pd.DataFrame(data={'x': x, 'y': y})
##f= plt.subplots(figsize=(7, 7))
#ax.set(xscale="log", yscale="log")
```
